```python
import pandas as pd
import numpy as np
```

```python
# Load the CSV files  with the ids into a dataframe
df_irradianceIDs = pd.read_csv("data/revised_newmexico_installations.csv")
```

```python
locations= []
for location in df_irradianceIDs['NSRDB loc']:
    locations.append(location)
```

```python
dfs=[]
for index in range(len(locations)):
    df=pd.read_csv(f"data/NRELinstallation_solar_irradiance/{locations[index]}_2022.csv")
    dfs.append(df)
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
df
```

```python
df_location_groups = df.groupby(df['Location ID']).sum()
df_sums = df_location_groups.loc[:,['DNI','GHI']]
df_sums
```

```python
df_merged_NM_solar = pd.merge(df_irradianceIDs,df_sums,left_on='NSRDB loc',right_on='Location ID')
df_merged_NM_solar
```

```python
df_merged_NM_solar = df_merged_NM_solar.rename(columns={'DNI':'2022 total DNI', 'GHI':'2022 total GHI', 'p_cap_ac':'AC Rating (GW)', 
                                                        'p_cap_dc':'DC Rating (GW)','p_state':'state','p_county':'county','p_area':'area (sq meters)',
                                                        'ylat':'latitude','xlong':'longitude','p_pwr_reg':'utility','p_name':'name','p_tech_pri': 'principal tech','p_axis':'axis type'})
df_merged_NM_solar.head()
```
